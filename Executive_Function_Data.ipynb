# Description of Effect of TV on Executive Function Example

This is based on a study by Lillard, Drell, Richey, Boguszewski, & Smith (2015).

In this study, the authors measured Executive Function of 90 six-year olds after they watched shows that had either fast, medium, or slow pacing and either fantastical or realistic content.

```python
import numpy as np
import pandas as pd
EFC= pd.read_csv('executiveFunctionData.csv')
```

RENAMING COLUMNS

```python
# To avoid confusion about what the DV is, col name 'DV' has been renamed 'Exec Function'
EFC= EFC.rename(columns= {'DV': 'Exec_Function'})
```

SUMMARY STATISTICS

```python
EFC.describe()
# GM= 17.90
```

ARE THE GROUP MEANS DIFFERENT FROM EACHOTHER?

```python
print(EFC.groupby('Content').mean('DV'))
print(EFC.groupby('Pacing').mean('DV'))
```

```python
# Mean for Realistic Content Data

FRdat= EFC.loc[(EFC.Content=='Realistic')]
print(FRdat.Exec_Function.mean().round(2))

# Mean for Fantastical Content Data
FFdat= EFC.loc[(EFC.Content=='Fantastical')]
print(FFdat.Exec_Function.mean().round(2))

# Comparing the exective functioning means by Content shows executive functioning
# for children watching realistic content is lower(m= 23.65) compared to the executive functioning
# for children watching fantastical content(m= 12.16)
```

```python
# Executive Functioning Means of Realistic Content by Pacing

FRF= FRdat.loc[(FRdat.Pacing=='Fast')]
print(FRF.describe().round(2))
# Faster response or Higher Executive functioning(m= 21.90)

FRM= FRdat.loc[(FRdat.Pacing=='Medium')]
print(FRM.describe().round(2))
# Medium response or intermediate Executive functioning(m= 24.09)

FRS= FRdat.loc[(FRdat.Pacing=='Slow')]
print(FRS.describe().round(2))
#Slower response or Lower Executive functioning(m= 24.96)
```

```python
# Executive Functioning Mean of Fantastical Content by Pacing

FFF= FFdat.loc[(FFdat.Pacing=='Fast')]
print(FFF.describe().round(2))

# The mean executive functioning/response time for Medium paced Fantastical content is lower(m= 10.19) compared to that of Fast paced Fantastical content(m=9.31)

FFM= FFdat.loc[(FFdat.Pacing== 'Medium')]
print(FFM.describe().round(2))

FFS= EFC.loc[(EFC.Pacing=='Slow')]
print(FFS.describe().round(2))
# The mean executive funtioning is lowest for children who watched slow paced fantastical content
```

## What does the difference in mean Executive Functioning for Content and Pacing Tell us?

The difference in mean executive functioning for realistic and fantastical content tells us there is a difference between the two groups. The mean difference in executive functions for pacing also tells us that there is a difference in executive functioning for slow, medium, and fast pacing.
